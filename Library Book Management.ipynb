```python
from datetime import datetime

tod=datetime.today()
formt=tod.strftime("Date:%d/%m/%Y\nTIME: %H:%M")

print(formt)
print("\n")

class Book():
    
    def __init__(self,title,author,genre,status):
        self.title=title
        self.author=author
        self.genre=genre
        self.status=status
    
    def __str__(self):
        return f'Title: {self.title}\nAuthor: {self.author}\nGenre: {self.genre}\n'

class Library():
    def __init__(self):
        self.books=[]
        self.read=[]
        self.notread=[]

    def add_book(self):
        try:
            number=int(input("Enter the Number of Books: "))
            for i in range(number):
                print(f"Book: {i+1}")
    
                t=input("Enter Title of Book: ").title()
                a=input(f"Enter Author of {t}: ").title()
                g=input("Enter Genre: ").title()
                s=input("Enter Status: Read for r\nNot read for nr ").lower()
                if s=="r":
                    c=Book(t,a,g,s)
                    self.books.append(c)
                    self.read.append(c)
                elif s=="nr":
                    c=Book(t,a,g,s)
                    self.books.append(c)
                    self.notread.append(c)
                else:
                    print("Invalid Status Input")
        except:
            print("Invalid Input!!")
            

    def viewbooks(self):
        if not self.books:
            print("No Books Recorded here :(")
        else:
            for j in self.books:
                    print(j)
    def view_read_books(self):
        if not self.read:
            print("No Books Recorded here :(")
        else:
            for k in self.read:
                print(k)

    def view_not_read_books(self):
        if not self.notread:
            print("No Books Recorded here :(")
        else:  
            for h in self.notread:
                print(h)
l=Library()
choice=''
while True:
    try:
        
        choice=int(input("\n1:Add Books\n2:ViewBooks\n3:Already Read Books\n4:View Books that you want to read:\n5:Search\n6:Clear Library\n7:Exit library\nEnter: "))
    except:
        print("Invalid input :(\nWE TAKE NUMBERS(1-7)")
        print("\n")
    if choice==1:
        print("\n")
        l.add_book()
        print("\n")
    elif choice==2:
        print("\n")
        print("---"*5)
        l.viewbooks()
        print("\n")
    elif choice==3:
        print("\n")
        print("---"*5)
        l.view_read_books()
        print("\n")
    elif choice==4:
        print("\n")
        print("---"*5)
        l.view_not_read_books()
        print("\n")

    elif choice==5:
        
        search=input("Enter the Book you want to search in Library: ").lower()
        flag=False
        for p in l.books:
            if p.title.lower() == search:
                flag=True
                break
        if flag==True:
            print(f"Book {search} exists ")
        else:
            print(f"Book: {search} NOT exists ")
        
        
    elif choice==6:
        l.books.clear()
        l.read.clear()
        l.notread.clear()
        print("\n")
        print("Your Library currently has no Books!!")
        print("\n")
    elif choice==7:
        
        print("Exiting Library... Visit Again :)")
        break    
    else:
        print("Invalid Input")
```
